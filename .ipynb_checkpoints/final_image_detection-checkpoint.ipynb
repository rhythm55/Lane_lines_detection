```python
#import packages 

#to plot the image
import matplotlib.pyplot as plt

#to process the image like reading and showing
import matplotlib.image as mpimg

#to apply various functions for detection of lane lines
import cv2

# for vertices
import numpy as np

import math
```

```python
# read the image
image = plt.imread('test_images/solidWhiteRight.jpg')
# display the image read in 'image' variable
plt.imshow(image)
image.shape
```

```python
# function to converte the image as gray so that we can find strength of pixels to find boundaries of an object
# to display the gray image returned by this function use plt.imshow(image,cmap='gray')
def grayscale(img):
    gray_image = cv2.cvtColor(img,cv2.COLOR_RGB2GRAY)
    return(gray_image)
```

```python
# this function is used to remove noise and smoothening the image 
def gaussian_blur(img,kernel_size):
    return cv2.GaussianBlur(img,(kernel_size,kernel_size),0)
```

```python
# this function is for canny edge detection to identify boundaries of an object
def canny(img,low_threshold,high_threshold):
    return cv2.Canny(img,low_threshold,high_threshold)
```

```python
# this function is used for masking the region that is blackning the region in which we are not interested
def region_of_interest(img):
    #design a blank mask
    # zeros_like is used to return array of zero with same size and type
    mask = np.zeros_like(img)
    #define channel to fill the mask
    if len(img.shape) > 2:
        channel_count = img.shape[2]
        ignore_mask_color = (255,) * channel_count
    else:
        ignore_mask_color = 255
    
    # define a four sided polygon to mask
    imshape = image.shape
    vertices = np.array([[(160,540),(410,320),(540,320),(840,540)]],dtype=np.int32)
    #(120,540),(400,310),(430,310),(200,540)],[(800,540),(580,310),(630,310),(830,540)
    cv2.fillPoly(mask, vertices, ignore_mask_color)
    
    #return the image with mask pixels which are nonzero
    # bitwise_and calculates conjunction of two arrrays
    masked_edges = cv2.bitwise_and(img, mask)
    
    return masked_edges
```

```python
# function used to detect the lines in image as we have to identify lane lines
def hough_lines(img,rho,theta,threshold,min_line_length,max_line_gap):
    lines = cv2.HoughLinesP(img, rho, theta, threshold, np.array([]), minLineLength=min_line_length, maxLineGap=max_line_gap)
    # create a image on which detected lines can be drawn
    line_img = np.zeros((img.shape[0], img.shape[1], 3), dtype=np.uint8)
    draw_lines(line_img, lines)
    return line_img
```

```python
# to find the detected lines are right lane or left lane line
def left_or_right(x1,y1,x2,y2):
    x,y=[],[]
    x += [x1,x2]
    y += [y1,y2]
    m_and_c = np.polyfit(x,y,1)
    m,c=m_and_c
    return(m,c)
```

```python
#to draw lines to the given end points found as hough lines
def draw_lines(img, lines, color=[255, 0, 0], thickness=15):
    left_lines = []
    right_lines = []
   
    for line in lines:
        for x1,y1,x2,y2 in line:
            m,c = left_or_right(x1,y1,x2,y2)
            if m<0:
                left_lines.append((m,c))
            else:
                right_lines.append((m,c))
    left_avg = np.average(left_lines,axis=0)
    right_avg = np.average(right_lines,axis=0)
    def calc_x(y,m,c):
        x = int((y-c)/m)
        return(x)
    cv2.line(img, (calc_x(img.shape[0],left_avg[0],left_avg[1]), img.shape[0]), (calc_x(320,left_avg[0],left_avg[1]),320), color, thickness) 
    cv2.line(img, (calc_x(img.shape[0],right_avg[0],right_avg[1]), img.shape[0]), (calc_x(320,right_avg[0],right_avg[1]),320), color, thickness)
        
```

```python
# calculates the weighted sum of two arrays
def weighted_img(img, edges, α=0.8, β=1., γ=0.):
    #color_edges = np.dstack((edges,edges,edges))
    return cv2.addWeighted(edges, α, img, β, γ)
```

```python
# now let us apply our pipeline to see if it works
# our test image is stored in variable image
#1:converte it in gray image
gray_image = grayscale(image)
#2.Remove the noises and smoothen your image
gauss_image = gaussian_blur(gray_image,19)
#3: apply canny to detect the edges
canny_image = canny(gauss_image,30,100)
#4:now we will select the region in which we are interested
interest_image = region_of_interest(canny_image)
#5: We will detect the line object in our image and plot it
hough_image = hough_lines(interest_image,2,np.pi/180,25,10,1)
#6: now we will combine both the images
final = weighted_img(hough_image,image)
plt.imshow(final)
```
