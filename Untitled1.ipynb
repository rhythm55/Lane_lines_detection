```python
# import all the libraries 
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import numpy as np
import cv2
```

```python
# read the image
image = mpimg.imread('test_images/solidWhiteRight.jpg')

gray = cv2.cvtColor(image,cv2.COLOR_RGB2GRAY)
#print the image type and dimensions
print("this image is : ",type(gray)," with dimensions:",gray.shape)
plt.imshow(gray,cmap='gray')
```

```python
#apply gaussian smoothning
kernel_size = 3
blur_gray = cv2.GaussianBlur(gray,(kernel_size,kernel_size),0)
plt.imshow(blur_gray,cmap = 'gray')
```

```python
# apply canny algorithm to detect high strength pixels
low_threshold = 50
high_threshold = 150
edge = cv2.Canny(blur_gray,low_threshold,high_threshold)
plt.imshow(edge,cmap = 'Greys_r')
```
